```python
import matplotlib.pyplot as plt

dpu_archs = [
    "B512", "B800", "B1024", "B1152", "B1600", "B2304", "B3136", "B4096"
]
performance = [
    1.2997713489634368,
    1.8472203741244175,
    2.213285954411619,
    2.4923075734652063,
    3.047977531769629,
    3.9782584426661227,
    4.769996643958069,
    5.950992292221635,
]

plt.figure(figsize=(10,6))
plt.bar(dpu_archs, performance, color='skyblue')
plt.xlabel('DPU Architecture')
plt.ylabel('Performance (FPS)')
plt.title('DPU Architecture vs Performance (FPS)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Performance relative to B512

```python
dpu_archs = [
    "B512", "B800", "B1024", "B1152", "B1600", "B2304", "B3136", "B4096"
]

performance = [
    1.2997713489634368,
    1.8472203741244175,
    2.213285954411619,
    2.4923075734652063,
    3.047977531769629,
    3.9782584426661227,
    4.769996643958069,
    5.950992292221635,
]

# Normalize performance to B512
relative_performance = [p / performance[0] for p in performance]

plt.figure(figsize=(10,6))
plt.bar(dpu_archs, relative_performance, color='mediumseagreen')
plt.xlabel('DPU Architecture')
plt.ylabel('Relative Performance (vs B512)')
plt.title('Relative Performance of DPUCZDX8G Architectures')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
